```python
def kesson_table(df):
    null_val = df.isnull().sum()
    percent = 100 * df.isnull().sum() / len(df)
    kesson_table = pd.concat([null_val,percent],axis=1)
    kesson_table_ren_colums = kesson_table.rename(columns= {0 : '欠損数',1:'%'})
    return kesson_table_ren_colums

kesson_table(train)
kesson_table(test)
```
